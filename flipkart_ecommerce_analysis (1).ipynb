```python
# Clean the Price column (remove  and commas, convert to numeric)
df['Price_clean'] = (
    df['Price']
    .str.replace('', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

# Get distribution statistics
df['Price_clean'].describe()
```

6.)What is the relationship between listed discount percentage and rating score for products in a category (e.g., do higher discounts tend to have lower ratings)?
 

```python
df['discount_clean'] = (
    df['Discount']
    .str.replace('% off', '', regex=False)
    .str.replace('%', '', regex=False)
    .fillna('0')          # handle missing values
    .astype(float)
)

df['discount_clean'].dtype
df[['discount_clean', 'Rating']].corr()


```

```python
Insight:
The relationship between listed discount percentage and rating score is generally weak to slightly negative. Products with higher discounts often tend to have slightly lower ratings, while highly rated products usually offer moderate or lower discounts.
```
